```python
# Import libraries
import csv
import pandas as pd
from tqdm.notebook import tqdm
import snscrape.modules.twitter as sntwitter
```

```python
path = "/home/jupyter/agra/source code/sentiment analytics/"
```

```python
# input keyword
keyword = '@Kopikenangan_ID OR Kopi Kenangan OR #FromMantanToUnicorn'
```

```python
# input date
scraper = sntwitter.TwitterSearchScraper(keyword + 'since:2022-01-01 until:2023-01-01')
```

```python
# start scraping data
tweets = []
maxTweets = 10_000
for i,tweet in tqdm(enumerate(scraper.get_items()), total=maxTweets) :
    data = [
        tweet.date,
        tweet.id, 
        tweet.rawContent, 
        tweet.user.username, 
        tweet.user.verified, 
        tweet.user.followersCount, 
        tweet.user.friendsCount, 
        tweet.mentionedUsers, 
        tweet.replyCount, 
        tweet.retweetCount, 
        tweet.likeCount, 
        tweet.media, 
        tweet.lang, 
        tweet.user.location
    ]
    
    tweets.append(data)
    
    if i > maxTweets:
        break
```

```python
# create df
tweet_df = pd.DataFrame(
    tweets,columns=['date', 
                    'tweet_id', 
                    'text', 
                    'username', 
                    'verified', 
                    'followers',
                    'following',
                    'mentioned users',
                    'retweet',
                    'like ', 
                    'reply', 
                    'media', 
                    'language', 
                    'location'] 
)                    
```

```python
# save to csv
tweet_df.to_csv(path + 'data/resto-scrape.csv', index = False)
```
